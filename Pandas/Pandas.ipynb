```python
df = pd.read_csv('houses_to_rent.csv')
df
```

```python
df_l = df.loc[:, ['city', 'total (R$)']]
df_l
```

```python
df_plot = df_l.groupby('city').agg('mean')
df_plot
```

```python
df_plot.plot(kind='bar',figsize=(20,10))
```

```python
df_plot.plot(kind="bar", figsize=(20,10))

plt.title('aluguéis') #Adicionando o titulo
plt.xlabel('UF')
plt.ylabel('Aluguel Total')
plt.show()
```

```python
sns.boxplot(y='total (R$)', x='city',
                   data=df_l,
                   width=0.5,
                   palette="colorblind")
```

```python
df_bh = df_l[(df_l['city'] == 'Belo Horizonte') & (df_l['total (R$)']<1000000)]
df_bh
```

```python
sns.boxplot(y='total (R$)', x='city',
                   data=df_bh,
                   width=0.5,
                   palette='colorblind')
```
